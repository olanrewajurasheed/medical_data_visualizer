```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('medical_examination.csv', delimiter = ',')
df
```

```python
df['overweight'] = (df['weight']/((df['height'] / 100) ** 2))
df['overweight'] = (df['overweight'].apply(lambda x: 1 if x > 25 else 0))
df = pd.DataFrame(df)
df.head(2)
```

```python
df['cholesterol'].max()
mapping = {1:0, 2:1, 3:1}
df['cholesterol'] = df['cholesterol'].map(mapping)
df['gluc']  = df['gluc'].map(mapping)
df
```

```python
df_cat = pd.DataFrame(df)
df_cat = pd.melt(df, value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'],
                 id_vars = ['cardio'])
df_cat
```

```python
df_cat = pd.DataFrame(df_cat.groupby(['variable', 'value', 'cardio'])['value'].count()).rename(columns = {'value': 'total'}).reset_index()
df_cat
```

```python
sns.catplot( x = 'variable', y = 'total', hue = 'value',
               col = 'cardio', data = df_cat, kind = 'bar')
plt.savefig('catplot.png')
```

```python
#data cleaning and filtering of wrong data
df_heat = pd.DataFrame(df)
df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
             (df['height'] >= df['height'].quantile(0.025)) &
             (df['height'] <= df['height'].quantile(0.975)) &
             (df['weight'] >= df['weight'].quantile(0.025)) &
             (df['weight'] <= df['weight'].quantile(0.975))]

#checking for correlation matrix after the error has been corrcted

corr = df_heat.corr()
corr
```

```python
#Generating a mask for uppper triangle

mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
mask
```

```python
#setting up matplotlib figures
fig, ax = plt.subplots(figsize=(11,9))
fig = sns.heatmap(corr, annot=True, mask = mask, fmt = '.1f', vmax = .3, center = 0,
          square = False, linewidths = .5, cbar_kws = {'shrink': .5})
```

```python
plt.savefig('heatmap.png')
```
